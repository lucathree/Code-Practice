### 레이블 인코딩

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
activity_train = train['Activity']
encoder.fit(activity_train)
labels_train = encoder.transform(activity_train)
labels_train
```

```python
train['labels'] = labels_train
train.iloc[:,-2:]
```

```python
activity_test = test['Activity']
encoder.fit(activity_test)
labels_test = encoder.transform(activity_test)
test['labels'] = labels_test
test.iloc[:,-2:]
```

```python
train['Activity'].unique()
```

```python
# 인코딩이 알파벳 순으로 되었기 때문에 타겟값의 번호가 섞여있다. 
train['labels'].unique()
```

### 데이터셋 분리

```python
x_train = train.iloc[:,:-2]
x_train
```

```python
y_train = train['labels']
y_train
```

```python
x_test = test.iloc[:,:-2]
y_test = test['labels']
```

# 결정 트리를 이용한 예측

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
dt_clf = DecisionTreeClassifier()
dt_clf.fit(x_train, y_train)
dt_pred = dt_clf.predict(x_test)
```

```python
print(f'결정트리 예측 정확도: {accuracy_score(y_test, dt_pred)}')
print(f'트리 깊이: {dt_clf.get_depth()}')
print(f'리프 노드 개수: {dt_clf.get_n_leaves()}')
```

### GridSearchCV 를 통한 교차검증 및 파라미터 수정

```python
from sklearn.model_selection import GridSearchCV
```

```python
dt_clf.get_params()
```

```python
dt_params = {
    'max_depth':[6,10,14,18,22], 
    'min_samples_split':[2,4,6]
}

grid_cv = GridSearchCV(dt_clf, param_grid=dt_params, scoring='accuracy', cv=4, verbose=1)
grid_cv.fit(x_train, y_train)
```

```python
print(f'그리드서치 후 최고 평균 정확도: {grid_cv.best_score_}')
print(f'최적 하이퍼 파라미터: {grid_cv.best_params_}')
```

```python
new_dt_clf = grid_cv.best_estimator_
best_dt_pred = new_dt_clf.predict(x_test)
print(f'하이퍼 파라미터 최적화 후 테스트 정확도: {accuracy_score(y_test, best_dt_pred)}')
```

```python
# 데이터프레임으로 결과 확인
result_df = pd.DataFrame(grid_cv.cv_results_)
result_df
```

### 결과 시각화

```python
import matplotlib.pyplot as plt
```

```python
fi_values = new_dt_clf.feature_importances_
fi = pd.Series(fi_values, index=x_train.columns)
fi
```

```python
# 중요도순 정렬
sort1 = fi.sort_values(ascending=False)[:20]
sort2 = sort1.sort_values()
sort2
```

```python
plt.figure(figsize=(8,6))
plt.title('feature importance')
plt.barh(sort2.index, sort2)
plt.show()
```

```python
# seaborn을 사용한 막대그래프
import seaborn as sns
sns.barplot(x=sort1, y=sort1.index)
plt.show()
```

```python
# 트리 시각화
import graphviz
from sklearn.tree import export_graphviz

export_graphviz(new_dt_clf,
                out_file='tree2.dot',
                max_depth=None,
                feature_names=x_train.columns,
                class_names=train['Activity'].unique(),
                filled=True)

with open('tree2.dot') as f:
  dot_graph = f.read()
graphviz.Source(dot_graph)
```
