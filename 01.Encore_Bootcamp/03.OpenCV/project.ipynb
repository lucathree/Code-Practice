```python
import sys, cv2
import numpy as np

cap = cv2.VideoCapture('img/mycats.avi')
fps = cap.get(5)
print(fps)

if not cap.isOpened():
    print('Video open failed!')
    sys.exit()


classifier = cv2.CascadeClassifier('haar/haarcascade_frontalcatface.xml')

if classifier.empty():
    print('XML load failed!')
    sys.exit()

   
while cap.isOpened(): # 동영상 파일이 정상 오픈이면
    ret, frame = cap.read() # 동영상에서 frame을 읽음
    if not ret:
        break
    catfaces = classifier.detectMultiScale(frame)
    
    for (x, y, w, h) in catfaces:
        cv2.rectangle(frame, (x, y, w, h), (255, 0, 255), 2)
        
    cv2.imshow('cctv', frame)  # 읽은 frame을 윈도우에 출력
    if cv2.waitKey(25) == 27: # esc입력 시 종료 
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)#카메라 오픈

if not cap.isOpened():# cap.isOpen():카메라 정상오픈이면 True, 아니면 False
    print('카메라 오픈 안됨')
    sys.exit(0)

classifier = cv2.CascadeClassifier('haar/haarcascade_frontalcatface.xml')

if classifier.empty():
    print('XML load failed!')
    sys.exit()

while True:
    ret, frame = cap.read() 
    if ret: #정상 읽기일 때만
        faces = classifier.detectMultiScale(frame)
        
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y, w, h), (255, 0, 255), 2)
        
    cv2.imshow('img', frame) #영상을 윈도우에 출력    
    if cv2.waitKey(30) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import sys
import numpy as np
import cv2

src = cv2.imread('img/mycats2.jpg')

if src is None:
    print('Image load failed!')
    sys.exit()

classifier = cv2.CascadeClassifier('haar/haarcascade_frontalcatface_extended.xml')

if classifier.empty():
    print('XML load failed!')
    sys.exit()

faces = classifier.detectMultiScale(src)

for (x, y, w, h) in faces:
    cv2.rectangle(src, (x, y, w, h), (255, 0, 255), 2)

cv2.imshow('src', src)
cv2.waitKey()
cv2.destroyAllWindows()
```
