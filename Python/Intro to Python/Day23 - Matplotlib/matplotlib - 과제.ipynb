## B-1. 회사명으로 정보 검색

```python
def search(name):
    try:
        res = df1.loc[name]
        #resdf = pd.DataFrame(res)
        return res
    except Exception:
        print('회사 정보가 없습니다')

search('없는회사')
search('휴벡셀')
```

**+ 검색어를 포함하는 모든 결과 출력하기**

```python
def search_like(name):
    list=[]
    for s in df1.index:
        if name in s:
            res = df1.loc[s]
            list.append(res)
    data = pd.DataFrame(list)
    if list == []:
        print('일치하는 회사가 없습니다')
    else:
        return data

search_like('없는회사')
search_like('거북선')
```

## B-2. 상장회사의 지역 분포도를 그래프로 표현

```python
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False

df3 = df1['지역'].dropna()
count = df3.value_counts()
val = count.values
idx = count.index
plt.bar(idx, val)

plt.xticks(rotation=90)
plt.xlabel('지역')
plt.ylabel('상장회사 수')
plt.title('상장회사 지역분포도')
plt.show()
```

## B-3. 상장연도별 분포도를 그래프로 표현

```python
df4 = df1.sort_values(by='상장일')
dates = df4['상장일']
dates
```

```python
idx = dates.index
years = []
for year in range(1956, 2021):
    for date in dates:
        if str(year) in str(date):
            years.append(year)
y = pd.DataFrame({'상장연도':years}, index=idx)
y
```

```python
plt.hist(years, bins='auto')
plt.title('상장연도별 분포도')
plt.xticks(range(1956,2021,4), rotation=45)
plt.show()
```
