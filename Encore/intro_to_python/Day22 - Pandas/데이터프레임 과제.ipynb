```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('도로교통공단_가해운전자 연령층별 월별 교통사고 통계.csv', encoding='euc-kr')
df
```

### 1. 연령별 데이터 출력

'가해자연령층' 컬럼 값을 배열로 반환

```python
a = df['가해자연령층'].values
v = np.unique(a)  # 중복되는 값들을 제거
v
```

'가해자연령층'을 상위컬럼으로 갖는 멀티컬럼 데이터프레임 생성.  
연령층별 데이터프레임을 생성한 후 join으로 통합

```python
df2 = pd.DataFrame(index = range(0,12), columns=[[],[]])  #연령별 데이터프레임을 담을 빈 데이터프레임 생성
for i in v:  #'가해자연령층'별 데이터프레임 생성 후 df2에 추가
    temp = df[df['가해자연령층']==i][df.columns[1:]]
    val=temp.values
    title = i
    cols=[[title,title,title,title,title,title], temp.columns]
    x = pd.DataFrame(val, columns=cols)
    df2 = df2.join(x)
df2
```

### 2. 연령별 최대 사고발생월과 사고수

```python
df3 = pd.DataFrame(columns=['월','사고건수'])  #빈 데이터프레임 생성
for i in v:
    temp = df2[i][['월','사고건수']]  #연령별 '월', '사고건수' 만 갖는 데이터프레임 생성
    val = temp.max()['사고건수']  #최대사고수 구하여 값을 val에 저장
    data = temp[temp['사고건수']==val]  #최대사고수를 가진 '월'과 '사고건수'를 담을 데이터프레임 생성
    df3 = df3.append(data)  #df3에 행 추가
df3
```

```python
df3['가해자연령'] = v  #'가해자연령'을 기준으로 하는 컬럼 추가
df3.set_index('가해자연령')  #'가해자연령'컬럼을 인덱스로 설정
```

### 3. 연령별 12개월의 사고발생 총건, 총 사망자수

```python
total = df2.sum()
total  #모든 컬럼별 총합을 담은 시리즈
```

```python
df4 = pd.DataFrame(index=['사고건수', '사망자수'], columns=[])  #빈 데이터프레임 생성
for i in v:
    s = total[i][['사고건수', '사망자수']]  #총합이 계산된 시리즈에서 '가해자연령'과 사고건수'와 '사망자수'를 멀티인덱스로 갖는 시리즈 반환
    s2 = s.to_frame(name=i)  #시리즈를 데이터프레임으로 변환
    df4 = df4.join(s2)  #df4에 신규 데이터프레임 통합
df4.rename_axis('총합')  #인덱스명 정의
```

### 4. 연령별 12개월 사고 발생 평균 값

```python
avg = df2.mean()  #모든 컬럼별 평균값을 담은 시리즈 생성
idx = ['사고건수', '사망자수', '중상자수', '경상자수', '부상신고자수']  #반복해서 사용될 인덱스값을 리스트로 저장하여 사용
df5 = pd.DataFrame(index=idx, columns=[])
for i in v:
    s = avg[i][idx]
    s2 = s.to_frame(name=i)
    df5 = df5.join(s2)
df5.round(2).rename_axis('평균')  #실수의 자리수를 소수점 2번째 자리까지로 반올림하고 인덱스명 '평균'으로 정의
```
