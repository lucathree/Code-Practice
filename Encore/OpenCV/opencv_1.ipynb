```python
import cv2
from matplotlib import pyplot as plt # as alias  

img = cv2.imread('a.jpg', cv2.IMREAD_COLOR)

#opencv   r,g,b  
b, g, r = cv2.split(img)   # img b,g,r 
img2 = cv2.merge([r,g,b]) # b,,g, r  Merge

plt.imshow(img2)
plt.xticks([]) # x 
plt.yticks([]) # y 
plt.show()
```
