# Voting Classifier
- naive bayes, decision tree, logistic regression, knn 사용
- 나이브 베이즈는 피처간 독립성을 전제로 하기 때문에 이 데이터셋에는 부적합 할 수 있음. 다만, voting classifier에 미치는 영향을 보기 위해서 사용

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
```

```python
nb_clf = GaussianNB()
dt_clf = DecisionTreeClassifier()
lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier()

vt_clf = VotingClassifier(estimators=[('nb', nb_clf), ('dt', dt_clf), ('lr', lr_clf), ('knn', knn_clf)])

x_train, x_test, y_train, y_test = train_test_split(df, target, test_size=0.3, shuffle=True, random_state=777, stratify=target)

vt_clf.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

def get_accuracy(model, x_train, y_train, x_test, y_test):
    print(f'분류기: {model}')
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    acc = accuracy_score(y_test, pred)
    print(f'정확도: {acc:.4f}')
    print()
    return acc
```

#### naive bayes 분류기를 포함한 결과

```python
accuracies = []
accuracies.append(get_accuracy(nb_clf, x_train, y_train, x_test, y_test))
accuracies.append(get_accuracy(dt_clf, x_train, y_train, x_test, y_test))
accuracies.append(get_accuracy(lr_clf, x_train, y_train, x_test, y_test))
accuracies.append(get_accuracy(knn_clf, x_train, y_train, x_test, y_test))

print(f'평균: {sum(accuracies) / len(accuracies):.4f}')
```

```python
get_accuracy(vt_clf, x_train, y_train, x_test, y_test)
```

#### naive bayes 분류기를 제외한 결과

```python
vt_clf2 = VotingClassifier(estimators=[('dt', dt_clf), ('lr', lr_clf), ('knn', knn_clf)])
get_accuracy(vt_clf2, x_train, y_train, x_test, y_test)
```

#### hard voting 대신 soft voting을 사용했을 때

```python
vt_clf3 = VotingClassifier(estimators=[('nb', nb_clf), ('dt', dt_clf), ('lr', lr_clf), ('knn', knn_clf)], voting='soft')
get_accuracy(vt_clf3, x_train, y_train, x_test, y_test)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rf_clf = RandomForestClassifier()
get_accuracy(rf_clf, x_train, y_train, x_test, y_test)
```

```python
rf_clf.get_params()
```

```python
rf_params = {
    'max_depth':[4,6,10,14,18],
    'min_samples_leaf':[1,5,9,13],
    'min_samples_split':[2,6,10,14],
    'n_estimators':[50,100,200]
}

cv = GridSearchCV(rf_clf, rf_params, cv=3, verbose=1)
cv.fit(x_train, y_train)
```

```python
cv.best_params_
```

```python
cv.best_score_
```

```python
result_df = pd.DataFrame(cv.cv_results_)
result_df.sort_values('rank_test_score').head(10)
```

```python
new_rf_clf = cv.best_estimator_
new_rf_pred = new_rf_clf.predict(x_test)
print(f'하이퍼 파라미터 튜닝 후 정확도: {accuracy_score(y_test, new_rf_pred)}')
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
```

```python
print(f'RF 정밀도: {precision_score(y_test, new_rf_pred)}')
print(f'RF 재현율: {recall_score(y_test, new_rf_pred)}')
print(f'RF f1 스코어: {f1_score(y_test, new_rf_pred)}')
print(f'RF AUC: {roc_auc_score(y_test, new_rf_pred)}')
```

# Gradient Boosting Machine (GBM)

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_clf = GradientBoostingClassifier()
get_accuracy(gb_clf, x_train, y_train, x_test, y_test)
```

```python
gb_clf.get_params()
```

```python
gb_params = {
    'learning_rate':[0.05, 0.1, 0.3],
    'max_depth':[3,10],
    'min_samples_leaf':[2,4,8],
    'n_estimators':[100, 200, 400]
}

gb_cv = GridSearchCV(gb_clf, gb_params, cv=2, verbose=1)
gb_cv.fit(x_train, y_train)
```

```python
gb_df = pd.DataFrame(gb_cv.cv_results_)
gb_df.sort_values('rank_test_score').head(10)
```

```python
new_gb_clf = gb_clf.set_params(min_samples_leaf=4, n_estimators=200)
new_gb_pred = new_gb_clf.predict(x_test)

print(f'RF 정확도: {accuracy_score(y_test, new_rf_pred)}')
print(f'RF 정밀도: {precision_score(y_test, new_rf_pred)}')
print(f'RF 재현율: {recall_score(y_test, new_rf_pred)}')
print(f'RF f1 스코어: {f1_score(y_test, new_rf_pred)}')
print(f'RF AUC: {roc_auc_score(y_test, new_rf_pred)}')
```
